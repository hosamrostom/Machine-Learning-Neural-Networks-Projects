```python
from sklearn.model_selection import cross_val_score, KFold
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier, ExtraTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier, ExtraTreesClassifier, VotingClassifier , StackingClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
import numpy as np
from sklearn.linear_model import LogisticRegression

X = xtrain
y = ytrain

clf = [
    KNeighborsClassifier(n_neighbors=70),
    DecisionTreeClassifier(),
    RandomForestClassifier(n_estimators=70),
    GaussianNB(),
    SVC(probability=True),
    ExtraTreeClassifier(),
    GradientBoostingClassifier(n_estimators=70, learning_rate=0.1, max_features=3, max_depth=3, random_state=10),
    AdaBoostClassifier(n_estimators=70),
    ExtraTreesClassifier(n_estimators=70)
]

k_fold = KFold(n_splits=10, shuffle=True, random_state=0)

def model_fit():
    scoring = 'accuracy'
    for i, model in enumerate(clf):
        scores = cross_val_score(model, X, y, cv=k_fold, n_jobs=-1, scoring=scoring)
        print(f"\nModel {i} ({model.__class__.__name__}) Fold Scores:")
      #  for j, score in enumerate(scores, 1):
      #      print(f"  Fold {j}: {round(score * 100, 2)}%")
        print(f"  ➤ Average Accuracy: {round(np.mean(scores) * 100, 2)}%")



model_fit()



```

```python

model = RandomForestClassifier(n_estimators=70)
model.fit(X, y)
importances = model.feature_importances_

print("\nFeature Importances:")
for feature, importance in zip(X.columns, importances):
    print(f"{feature}: {importance:.4f}")

```

```python
estimators = [(f'model{i}', model) for i, model in enumerate(clf)]

voting_clf = VotingClassifier(estimators=estimators, voting='hard')

voting_clf.fit(X, y)

prediction = voting_clf.predict(xtest)

voting_score = cross_val_score(voting_clf, X, y, cv=k_fold, scoring='accuracy')
print(f"\n Hard Voting Accuracy: {model.__class__.__name__}) {round(np.mean(voting_score) * 100, 2)}%")




```

```python

stacking_clf = StackingClassifier(
    estimators=estimators,
    final_estimator=LogisticRegression(),
    cv=5,
    n_jobs=-1
)

stacking_clf.fit(X, y)
stacking_score = cross_val_score(stacking_clf, X, y, cv=k_fold, scoring='accuracy')
print(f"\n Stacking Classifier Accuracy: {round(np.mean(stacking_score) * 100, 2)}%")
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score

param_grid = {
    'KNeighborsClassifier': {
        'n_neighbors': [3, 5, 7, 9, 11]
    },
    'DecisionTreeClassifier': {
        'max_depth': [3, 5, 10, None],
        'criterion': ['gini', 'entropy']
    },
    'RandomForestClassifier': {
        'n_estimators': [50, 100],
        'max_depth': [5, 10, None]
    },
    'GaussianNB': {},
    'SVC': {
        'C': [0.1, 1, 10],
        'kernel': ['linear', 'rbf']
    },
    'ExtraTreeClassifier': {
        'max_depth': [3, 5, 10, None]
    },
    'GradientBoostingClassifier': {
        'n_estimators': [50, 100],
        'learning_rate': [0.01, 0.1],
        'max_depth': [3, 5]
    },
    'AdaBoostClassifier': {
        'n_estimators': [50, 100],
        'learning_rate': [0.5, 1.0]
    },
    'ExtraTreesClassifier': {
        'n_estimators': [50, 100],
        'max_depth': [5, 10, None]
    }
}

for model in clf:
    model_name = model.__class__.__name__
    print(f"\nTraining {model_name}...")

    if model_name in param_grid and param_grid[model_name]:
        grid = GridSearchCV(model, param_grid[model_name], cv=5, scoring='accuracy')
        grid.fit(xtrain, ytrain)
        best_model = grid.best_estimator_
        print(f"Best Params for {model_name}: {grid.best_params_}")
    else:
        best_model = model
        best_model.fit(xtrain, ytrain)

    preds = best_model.predict(xtest)
    acc = accuracy_score(ytest, preds)
    print(f"{model_name} Accuracy: {acc:.4f}")
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
import matplotlib.pyplot as plt

model = Sequential()
model.add(Dense(128, input_dim=xtrain.shape[1], activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.4))

model.add(Dense(64, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.3))

model.add(Dense(32, activation='relu'))
model.add(Dropout(0.2))

model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
checkpoint = ModelCheckpoint('best_model.keras', monitor='val_loss', save_best_only=True)

history = model.fit(
    X, y,
    epochs=200,
    batch_size=32,
    validation_split=0.2,
    callbacks=[early_stop, checkpoint],
    verbose=1
)

```

```python
plt.figure(figsize=(10,5))
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Val Accuracy')
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title("Model Performance")
plt.xlabel("Epochs")
plt.ylabel("Accuracy / Loss")
plt.legend()
plt.grid(True)
plt.show()

loss, acc = model.evaluate(xtest, ytest)
print(f"Test Accuracy: {acc*100:.2f}%")
```
